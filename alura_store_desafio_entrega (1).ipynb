# Análisis de Ventas - Alura Store
**Desafío Data Science LATAM**

Objetivo: Analizar el rendimiento de las cuatro tiendas de Alura Store para recomendar cuál debería vender el Sr. Juan para invertir en un nuevo emprendimiento.

## 1. Importar bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## 2. Cargar los datos

```python
tienda1 = pd.read_csv('/content/tienda_1 .csv')
tienda2 = pd.read_csv('/content/tienda_2.csv')
tienda3 = pd.read_csv('/content/tienda_3.csv')
tienda4 = pd.read_csv('/content/tienda_4.csv')

tienda1.head()
```

## 3. Ingresos por tienda

```python
ingresos = {
    'Tienda 1': tienda1['Precio'].sum(),
    'Tienda 2': tienda2['Precio'].sum(),
    'Tienda 3': tienda3['Precio'].sum(),
    'Tienda 4': tienda4['Precio'].sum()
}

ingresos
```

## 4. Gráfico de ingresos por tienda

```python
plt.figure()
plt.bar(ingresos.keys(), ingresos.values())
plt.title('Ingresos por tienda')
plt.xlabel('Tiendas')
plt.ylabel('Ingresos')
plt.show()
```

## 5. Promedio de calificaciones

```python
reseñas = {
    'Tienda 1': tienda1['Calificación'].mean(),
    'Tienda 2': tienda2['Calificación'].mean(),
    'Tienda 3': tienda3['Calificación'].mean(),
    'Tienda 4': tienda4['Calificación'].mean()
}

reseñas
```

```python
plt.figure()
plt.bar(reseñas.keys(), reseñas.values())
plt.title('Promedio de calificaciones por tienda')
plt.ylabel('Calificación')
plt.show()
```

## 6. Costo promedio de envío

```python
envio = {
    'Tienda 1': tienda1['Costo de envío'].mean(),
    'Tienda 2': tienda2['Costo de envío'].mean(),
    'Tienda 3': tienda3['Costo de envío'].mean(),
    'Tienda 4': tienda4['Costo de envío'].mean()
}

envio
```

```python
plt.figure()
plt.bar(envio.keys(), envio.values())
plt.title('Costo promedio de envío por tienda')
plt.ylabel('Costo de envío')
plt.show()
```

## 7. Categorías más vendidas

```python
categoria_tienda1 = tienda1['Categoría del Producto'].value_counts()

plt.figure()
categoria_tienda1.plot(kind='bar')
plt.title('Categorías más vendidas - Tienda 1')
plt.ylabel('Cantidad de ventas')
plt.show()
```

## 8. Conclusión

Después de analizar los ingresos, las calificaciones de los clientes y los costos de envío de las cuatro tiendas, se observa que **la Tienda 4 presenta el menor ingreso total**, lo que indica un rendimiento inferior en comparación con las demás tiendas.

Además, su desempeño general en las métricas analizadas sugiere que es la tienda menos eficiente dentro de la cadena.

**Por esta razón, se recomienda que el Sr. Juan considere vender la Tienda 4**, ya que representa el menor rendimiento y su venta permitiría liberar recursos para invertir en un nuevo emprendimiento más rentable.
